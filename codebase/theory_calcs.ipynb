```python
import numpy as np
```

### Stage 1: Passive band-pass filter

```python
# Capacitors
C_hp = 100e-9  # F
C_lp = 1e-9    # F


def corner_frequency(R, C):
    return 1 / (2 * np.pi * R * C)


# Cutoff frequencies
f_hp = corner_frequency(R=1e3, C=C_hp)
f_lp = corner_frequency(R=10e3, C=C_lp)

print(f"High-pass cutoff frequency: {f_hp:.2f} Hz")
print(f"Low-pass cutoff frequency:  {f_lp:.2f} Hz")
```

### Stage 5: Passive low-pass filter

```python
# RC time constant
R = 10e+3
C = 100e-9

tau = R * C

print(f"Time constant: {tau:.2e} s")
```
